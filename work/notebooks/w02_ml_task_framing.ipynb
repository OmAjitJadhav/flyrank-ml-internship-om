## My lane as an ML task

This is a **ranking** problem.

The model ranks web pages based on search performance, engagement metrics, and content freshness. Pages with the highest scores should be refreshed first because they have the greatest potential for improvement.

## 2. Target or proxy

## Target or proxy

The target is a content opportunity score that ranks web pages by their need for content refresh.

The score is based on observed data such as search performance, engagement metrics, and content freshness. It is created from measured website data rather than a fixed rule.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The success metric is NDCG (Normalized Discounted Cumulative Gain).

A higher NDCG score means the model ranks the most important pages near the top. Higher values indicate better ranking quality.

## 5. Why ML beats a fixed rule here

## Why ML beats a fixed rule here

A fixed rule cannot consider all the factors at the same time. Search performance, engagement metrics, content freshness, and competition all affect the decision.

Machine learning can learn patterns from historical data and rank pages more accurately than a single if-else rule. This helps teams focus on the pages with the highest opportunity for improvement.
